```python
# https://www.youtube.com/watch?v=pKz1faPVNMA&list=PL5eJgcQ87sgfUrc6an-0Swam0kQXVYnuw&index=16
from bs4 import BeautifulSoup
```

```python
import urllib.request
```

```python
router_technology_news = 'http://feeds.reuters.com/reuters/technologyNews'
router_technology_news_video = 'http://feeds.reuters.com/reuters/USVideoTechnology'
```

```python
req = urllib.request.urlopen(router_technology_news_video)
```

```python
req_xml = BeautifulSoup(req, 'xml')
```

```python
req_xml
```

```python
allItems = req_xml.findAll('item')
allItems
```

```python
for eachItem in allItems:
    print(eachItem)
```

```python
for eachItem in req_xml.findAll('title'):
    print(eachItem)
```

```python
# to reference without the tags
# use text
for eachItem in req_xml.findAll('title'):
    print(eachItem.text)
```

```python
# to reference without the tags
# use text
for eachItem in req_xml.findAll('link'):
    print(eachItem.text)
```
