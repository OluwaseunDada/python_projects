```python
# https://www.youtube.com/watch?v=pKz1faPVNMA&list=PL5eJgcQ87sgfUrc6an-0Swam0kQXVYnuw&index=16
# https://blog.feedspot.com/world_news_rss_feeds/
from bs4 import BeautifulSoup
```

```python
import urllib.request
```

```python
npr_world_news_rss = 'https://www.npr.org/rss/rss.php?id=1004'
```

```python
req = urllib.request.urlopen(npr_world_news_rss)
```

```python
req_xml = BeautifulSoup(req, 'xml')
```

```python
req_xml
```

```python
allItems = req_xml.findAll('item')
allItems
```

```python
for eachItem in allItems:
    print(eachItem)
```

```python
for eachItem in req_xml.findAll('title'):
    print(eachItem)
```

```python
# to reference without the tags
# use text
for eachItem in req_xml.findAll('title'):
    print(eachItem.text)
```

```python
# to reference without the tags
# use text
for eachItem in req_xml.findAll('link'):
    print(eachItem.text)
```
